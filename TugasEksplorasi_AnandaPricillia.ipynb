```
import pandas as pd # import library to read data into dataframe
url = 'https://raw.githubusercontent.com/anandapricillia/EksplorasiData/master/cereal.csv'
```

```
import pandas as pd
snack = pd.read_csv(url, error_bad_lines=False)
snack.head()
```

```
import matplotlib.pyplot as plt
import pandas as pd
snack = pd.read_csv(url, error_bad_lines=False)

#Plots in matplotlib reside within a figure
fig=plt.figure()

ax = fig.add_subplot(1,1,1)

#variabel
ax.hist(snack['calories'],bins = 10)

#Labels and Tit
plt.title('calories of the cereal')
plt.xlabel('calories')
plt.ylabel('jumlah kalori')
plt.show()
```

```
fig=plt.figure()

ax = fig.add_subplot(1,1,1)

#variabel
ax.scatter(snack['calories'],snack['protein'])

#Labels and Tit
plt.title('calories and protein')
plt.xlabel('calories')
plt.ylabel('protein')
plt.show()
```
